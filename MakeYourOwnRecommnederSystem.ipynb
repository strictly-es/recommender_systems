```python
#協調フィルタリング
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD

ratings = pd.read_csv("./ml_data/recommnder_data/goodbooks-10k/ratings.csv")
columns = ['book_id','user_id','rating']
```

```python
books = pd.read_csv("./ml_data/recommnder_data/goodbooks-10k/books.csv")

columns = ['id','book_id', 'best_book_id', 'work_id', 'books_count', 'isbn', 'isbn13', 'authors',
           'original_publication_year', 'original_title', 'title', 'language_code', 'average_rating',
           'ratings_count', 'work_ratings_count', 'work_text_reviews_count',
           'ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5', 'image_url', 'small_image_url']
```

```python
combined_books_data = pd.merge(ratings,books,on='book_id')
```

```python
combined_books_data.head()
```

```python
rating_utility_matrix = combined_books_data.pivot_table(values="rating",index="user_id",columns="title",fill_value=0)
```

```python
rating_utility_matrix.head()
```

```python
rating_utility_matrix.shape
```

```python
X = rating_utility_matrix.T
SVD = TruncatedSVD(n_components=30)
transposed_matrix = SVD.fit_transform(X)
transposed_matrix.shape
```

```python
corr_matrix = np.corrcoef(transposed_matrix)
books = rating_utility_matrix.columns

books_list = list(books)
book_1984 = books_list.index("1984")
```

```python
book_1984
```

```python
corr_1984 = corr_matrix[book_1984]
list(books[(corr_1984 < 1.0) & (corr_1984 > 0.8)])
```

```python
list(books[(corr_1984 < 1.0) & (corr_1984 > 0.7)])
```

```python
#コンテンツベース レコメンド2

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv("./ml_data/movie_dataset.csv")
df.head()
```

```python
features = ['keywords','cast','genres','director']


def combine_features(row):
    return row['keywords'] + " " + row['cast'] + " " + row["genres"] + " " + row["director"]

for feature in features:
    df[feature] = df[feature].fillna('')
    
df["combine_features"] = df.apply(combine_features,axis=1)

cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combine_features"])
```

```python
cosine_sim = cosine_similarity(count_matrix)

def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]

def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]

movie_user_likes = "Avatar"


movie_index = get_index_from_title(movie_user_likes)
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True )[1:]
```

```python
i=0
print("Top 5 similar movies to "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>5:
        break
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error

df = pd.read_csv('./ml_data/recommnder_data/melbourne-housing-market/Melbourne_housing_FULL.csv')
df.head()
```

```python

del df["Address"]
del df["Method"]
del df["SellerG"]
del df["Date"]
del df["Postcode"]
del df["Lattitude"]
del df["Longtitude"]
del df["Regionname"]
del df["Propertycount"]
```

```python
df.dropna(axis=0,how="any",thresh=None,subset=None,inplace=True)
```

```python
df = pd.get_dummies(df,columns=["Suburb","CouncilArea","Type"])
```

```python
X = df.drop("Price",axis=1)
y = df["Price"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,shuffle=True)
```

```python
model = ensemble.GradientBoostingRegressor(
    n_estimators=250,
    learning_rate=0.1,
    max_depth=5,
    min_samples_split=4,
    min_samples_leaf=6,
    max_features=0.6,
    loss=  "huber"
     )
model.fit(X_train,y_train)
```

```python
mae_test = mean_absolute_error(y_train,model.predict(X_train))
print("Training Set Mean Abosolute Error:%.2f"% mae_test)
```

```python
mae_test = mean_absolute_error(y_test,model.predict(X_test))
print("Training Set Mean Abosolute Error:%.2f"% mae_test)
```
